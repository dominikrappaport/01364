```python
from matplotlib.pyplot import plot
```

```python
from numpy.random import binomial
```

```python
def game(l, k, debug=False):
    g = [1]

    for i in range(k):
        g.append((1-l)*g[i] + l*2*binomial(1, 0.5)*g[i])

        if debug:
            print(f"Aktueller Spielstand: {g[i]}, Einsatz: {l*g[i]}, Saldo: {g[i+1]-g[i]}, Neuer Spielstand: {g[i+1]}")

    return g
```

```python
plot(game(1/4, 1000))
```

```python
plot(game(1/2, 1000))
```

```python
plot(game(3/4, 1000))
```

```python
plot(game(1, 1000))
```

```python
game(1/2, 10, True)
```
